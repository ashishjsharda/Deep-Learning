```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/(1+np.exp(-x))

x=np.array([-4,-3,-2,-1,0,1,2,3,4])
y=sigmoid(x)
print("Input values of x")
print(x)
print("\n Output Values (sigmoid(x)):")
print(y)
plt.xlabel("X")
plt.ylabel("Sigmoid(X)")
plt.plot(x,y)
plt.show()
```
